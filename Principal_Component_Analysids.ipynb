```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
filepath = '/content/iris.csv'
df_iris = pd.read_csv(filepath)
df_iris.head()
```

pca

```python
#split data into features and target
X= df_iris.drop(columns=['species'],axis=1) #features
y= df_iris['species'] #target
X.head()
```

```python
pca = PCA()
X_new = pca.fit_transform(X)
print(X_new[:5]) # Display the first 5 rows using slicing
```

```python
pca.get_covariance()
```

```python
explained_varience = pca.explained_variance_ratio_
print(explained_varience)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(range(1,len(explained_varience)+1), explained_varience) #alpha =0
plt.xlabel('Principal Components')
plt.ylabel('Explained Varience')
plt.show()
```

```python
pca_2 = PCA(n_components=3)
X_new_2 = pca_2.fit_transform(X)
explained_varience_2 = pca_2.explained_variance_ratio_
print('cummulative sum of explained varience ratio',sum(explained_varience_2))
```

```python
#new dataframe
# iris data has been transformed using only 2 principal components
df_iris_new = pd.DataFrame(X_new_2, columns=['PC1','PC2'])
df_iris_new['species'] = y
df_iris_new.head(3)
```
